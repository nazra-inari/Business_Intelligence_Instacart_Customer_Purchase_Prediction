## Data Dictionary
There are 6 separate tables that make up our data.  
1) orders  
2) products  
3) aisles  
4) departments  
5) order_products_prior  
6) order_products_train  

Below is the definition of the columns within those tables

`orders` (3.4m rows, 206k users):
* `order_id`: order identifier
* `user_id`: customer identifier
* `eval_set`: which evaluation set this order belongs in (see `SET` described below)
* `order_number`: the order sequence number for this user (1 = first, n = nth)
* `order_dow`: the day of the week the order was placed on
* `order_hour_of_day`: the hour of the day the order was placed on
* `days_since_prior`: days since the last order, capped at 30 (with NAs for `order_number` = 1)

`products` (50k rows):
* `product_id`: product identifier
* `product_name`: name of the product
* `aisle_id`: foreign key
* `department_id`: foreign key

`aisles` (134 rows):
* `aisle_id`: aisle identifier
* `aisle`: the name of the aisle

`deptartments` (21 rows):
* `department_id`: department identifier
* `department`: the name of the department

`order_products__SET` (30m+ rows):
* `order_id`: foreign key
* `product_id`: foreign key
* `add_to_cart_order`: order in which each product was added to cart (1 means it was the first item added to cart in that orderid)
* `reordered`: 1 if this product has been ordered by this user in the past, 0 otherwise

where `SET` is one of the four following evaluation sets (`eval_set` in `orders`):
* `"prior"`: orders prior to that users most recent order (~3.2m orders)
* `"train"`: training data supplied to participants (~131k orders)
* `"test"`: test data reserved for machine learning competitions (~75k orders)

Based on the data dictionary, we can understand that

# Data Engineering
It seems that missing from our data is a table for users. Considering that our primary research question is trying to predict the behavior of a user (i.e. whether or not they will re_order certain items), I think it is imperative to create a table for users. We will use this same table to populate features that we engineer for each user. 

```python
users = pd.DataFrame(orders_non_test['user_id'].unique(), columns=['user_id'])
users.head()
```

Furthermore, since the data is distributed amongst many tables, I foresee many instances of inner and left joins which are computationally expensive task, and hence I'm gonna create a master table with all the relevant data combined into a single dataframe. The downside to this being that this dataframe will have redundant information with repeating values. This is why data is usually stored in multiple tables with and entity relationship between the tables.

```python
a1 = pd.merge(orders_prior, order_products_prior, on = 'order_id', how = 'left')
a2 = pd.merge(a1, products, on = 'product_id', how = 'left')
a3 = pd.merge(a2, aisles, on = 'aisle_id', how = 'left')
df_master = pd.merge(a3, departments, on = 'department_id', how = 'left')
df_master.head()
```

Let's start adding some features to the users table

## Feature Engineering

### Count of Number Orders per User
The total number of orders (prior) that each user has.

```python
df_order_count_per_user = df_master.groupby(['user_id'])['order_id'].nunique().to_frame('counts')
df_order_count_per_user.reset_index(inplace = True)
df_order_count_per_user.head()
```

```python
df_order_count_per_user = orders_prior.groupby('user_id')['user_id'].count().to_frame('number_of_orders')
df_order_count_per_user.reset_index(inplace = True)
users = pd.merge(users, df_order_count_per_user, on = 'user_id', how = 'inner')
users.head()
```

### Number of Products Ordered by each User (prior) 
The total number of products ordered by each user prior to their most recent order.This is feature is not showing you the total number of unique products ordered. 

```python
products_per_order = order_products_prior.groupby('order_id')['order_id'].count().to_frame('number_of_products_ordered')
products_per_order.reset_index(inplace = True)
products_per_order.head()
```

```python
products_per_order = pd.merge(products_per_order, orders_prior, on = 'order_id', how = 'left')
products_per_user = products_per_order.groupby('user_id')['number_of_products_ordered'].sum().to_frame('total_products_ordered_prior')
products_per_user.reset_index(inplace = True)
users = pd.merge(users, products_per_user, on = 'user_id', how = 'left')
users.head()
```

### Number of Unique Products ordered (prior)
The number unique products ordered by the user over his/her entire shopping history with instacart

```python
order_products_prior.head()
```

```python
unique_products_per_user = df_master.groupby('user_id')['product_id'].nunique().to_frame('number_of_unique_products_ordered')
unique_products_per_user.reset_index(inplace = True)
users = pd.merge(users, unique_products_per_user, on = 'user_id', how = 'left')
users.head()
```

### Number of Products Re-Ordered
This feature represents the total number of products ordered that were re-orders by the user after the first order.

```python
products_reordered_per_user = df_master.groupby('user_id')['reordered'].sum().to_frame('total_products_reordered')
products_reordered_per_user.reset_index(inplace = True)
users = pd.merge(users, products_reordered_per_user, on = 'user_id', how = 'left')
users.head()
```

### Number of Unique Products Re-Ordered at least once
This feature represents the total number of unique products that were re-ordered by the user after the first order.

```python
unique_products_reordered_per_user = df_master[df_master['reordered'] == 1].groupby('user_id')['product_id'].nunique().to_frame('number_of_unique_products_reordered')
unique_products_reordered_per_user.reset_index(inplace = True)
users = pd.merge(users, unique_products_reordered_per_user, on = 'user_id', how = 'left')
users.head()
```

### Mean Number of Products per Order
This feature represents the average number of products in each order

```python
users['avg_products_per_order'] = users['total_products_ordered_prior'] / users['number_of_orders']
users.head()
```

### Mode Number of Products per Order
What is the mode (or most frequent) number of products the user orders per order

```python
mode_products_per_order = products_per_order.groupby('user_id')['number_of_products_ordered'].agg(pd.Series.mode).to_frame('mode_products_ordered')
mode_products_per_order.reset_index(inplace = True)
mode_products_per_order['mode_products_ordered'] = mode_products_per_order['mode_products_ordered'].map(lambda x: x.mean())
users = pd.merge(users, mode_products_per_order, on = 'user_id', how = 'left')
users.head()
```

### Standard Deviation of Number of Products per Order
This feature will represetnt the statndard deviation of the number of products ordered a user in a single order

```python
std_dev_products_per_order = products_per_order.groupby('user_id')['number_of_products_ordered'].std().to_frame('std_dev_products_ordered')
std_dev_products_per_order.reset_index(inplace = True)
users = pd.merge(users, std_dev_products_per_order, on = 'user_id', how = 'left')
users.head()
```

### Percentage of products re-ordered
The percentage of the unique products - ordered by the user - that were re-ordered

```python
users['percent_unique_products_reordered'] = users['number_of_unique_products_reordered'] / users['number_of_unique_products_ordered']
users.head()
```

### Top 3 Most Frequently bought Product ids and the counts
Engineering 6 features in total. The top 3 Most frequently bought product ids and the respective counts representing how many times each product was bough. A value of NaN in any of the features suggests that the user had bought less than 3 items. So for instance if the the 3rd most frequently bought product id and count is NaN, it implies that the user only ever purchased 2 items from insta-cart

```python
frequency_of_products_ordered = df_master.groupby(['user_id', 'product_id'])['product_id'].count().to_frame('count')
frequency_of_products_ordered.reset_index(inplace = True)
frequency_of_products_ordered = frequency_of_products_ordered.groupby(['user_id'])['product_id','count'].apply(lambda grp: grp.nlargest(n = 3, columns = 'count'))
frequency_of_products_ordered.reset_index(inplace = True)
frequency_of_products_ordered.head()
```

```python
f1 = frequency_of_products_ordered.groupby('user_id')['user_id'].count().to_frame('count')
f1.reset_index(inplace = True)
# Users that have ordered atleast 3 items
user_ids_3 = f1[f1['count'] == 3]['user_id']

# Users that have ordered only 2 items
user_ids_2 = f1[f1['count'] == 2]['user_id']

# Users that have ordered only 1 item
user_ids_1 = f1[f1['count'] == 1]['user_id']
```

```python
frequency_of_products_ordered.head()
```

```python
#Isolate the users by number of items
frequency_of_products_ordered_3items = frequency_of_products_ordered[frequency_of_products_ordered['user_id'].isin(user_ids_3)]
frequency_of_products_ordered_2items = frequency_of_products_ordered[frequency_of_products_ordered['user_id'].isin(user_ids_2)]
frequency_of_products_ordered_1items = frequency_of_products_ordered[frequency_of_products_ordered['user_id'].isin(user_ids_1)]

# Add product_id and product_count column names
frequency_of_products_ordered_3items['product_id_column_names'] = ['num_1_most_frequent_product_id', 'num_2_most_frequent_product_id', 'num_3_most_frequent_product_id']*(len(user_ids_3))
frequency_of_products_ordered_3items['product_count_column_names'] = ['num_1_most_frequent_product_qty', 'num_2_most_frequent_product_qty', 'num_3_most_frequent_product_qty']*(len(user_ids_3))

frequency_of_products_ordered_2items['product_id_column_names'] = ['num_1_most_frequent_product_id', 'num_2_most_frequent_product_id']*(len(user_ids_2))
frequency_of_products_ordered_2items['product_count_column_names'] = ['num_1_most_frequent_product_qty', 'num_2_most_frequent_product_qty']*(len(user_ids_2))

frequency_of_products_ordered_1items['product_id_column_names'] = ['num_1_most_frequent_product_id']*(len(user_ids_1))
frequency_of_products_ordered_1items['product_count_column_names'] = ['num_1_most_frequent_product_qty']*(len(user_ids_1))


# Drop unwanted columns
frequency_of_products_ordered_3items = frequency_of_products_ordered_3items.drop([ 'product_id', 'count'], axis = 1)
frequency_of_products_ordered_2items = frequency_of_products_ordered_2items.drop([ 'product_id', 'count'], axis = 1)
frequency_of_products_ordered_1items = frequency_of_products_ordered_1items.drop([ 'product_id', 'count'], axis = 1)

# Combine the three data frames
frequency_of_products_ordered_allitems = pd.concat([frequency_of_products_ordered_3items, 
                                                    frequency_of_products_ordered_2items,
                                                    frequency_of_products_ordered_1items])


# Left join with frequency_of_products_ordered dataframe to add column names to appropriate user ids
frequency_of_products_ordered = pd.merge(frequency_of_products_ordered,
                                         frequency_of_products_ordered_allitems,
                                         on = ['user_id', 'level_1'], how = 'left')

frequency_of_products_ordered.head()
```

### Generate Features based on Aisles and Departments

Let's throw in the feature generation code into a function so that we can avoid repetition. 

```python
def feature_generator(grouping_variable, feature_name_modifier, master_dataframe, users_dataframe):
    
    # total_unique_aisles_ordered_from
    unique_grouping_variables_ordered_from = (master_dataframe.groupby('user_id')[grouping_variable]
                                             .nunique()
                                             .reset_index()
                                             .rename(columns={grouping_variable:'total_unique_'+feature_name_modifier+'_ordered_from'}))

    # total_unique_aisles_reordered_from
    unique_grouping_variables_reordered_from = (master_dataframe[master_dataframe['reordered'] == 1].groupby('user_id')[grouping_variable]
                                                .nunique()
                                                .reset_index()
                                                .rename(columns={grouping_variable:'total_unique_'+feature_name_modifier+'_reordered_from'}))

    # Mean number of aisles ordered from in each order
    avg_unique_grouping_variables_ordered_from = (master_dataframe.groupby(['user_id', 'order_id'])[grouping_variable]
                                                  .nunique()
                                                  .reset_index()
                                                  .groupby('user_id')
                                                  .mean()
                                                  .reset_index()
                                                  .drop('order_id', axis = 1)
                                                  .rename(columns={grouping_variable:'avg_unique_'+feature_name_modifier+'_ordered_from'}))

    # Std Dev of Aisles ordered from in each order
    std_unique_grouping_variable_ordered_from = (master_dataframe.groupby(['user_id',  'order_id'])[grouping_variable]
                                                  .nunique()
                                                  .reset_index()
                                                  .groupby('user_id')
                                                  .std()
                                                  .reset_index()
                                                  .drop('order_id', axis = 1)
                                                  .rename(columns={grouping_variable:'std_unique_'+feature_name_modifier+'_ordered_from'}))


    # Mode unique aisles ordered from 
    mode_unique_grouping_variable_ordered_from = pd.concat([ users_dataframe[['user_id']],
                                                           (master_dataframe.groupby(['user_id', 'order_id'])[grouping_variable]
                                                             .nunique()
                                                             .to_frame('count')
                                                             .reset_index()
                                                             .groupby('user_id')['count']
                                                             .agg(pd.Series.mode)
                                                             .reset_index()
                                                             ['count']
                                                             .apply(lambda x: x.mean())
                                                             .to_frame('mode_unique_'+feature_name_modifier+'_ordered_from'))], axis = 1)

    list_of_feature_dfs = [unique_grouping_variables_ordered_from,
                           unique_grouping_variables_reordered_from,
                           avg_unique_grouping_variables_ordered_from,
                           std_unique_grouping_variable_ordered_from,
                           mode_unique_grouping_variable_ordered_from]


    dfs_merged = reduce(lambda  left,right: pd.merge(left,right,on='user_id',
                                                how='outer'), list_of_feature_dfs)
    
    return (dfs_merged)
```

```python
# Generate the lifetime features for aisles and departments
result_1_aisles = feature_generator(grouping_variable = 'aisle_id',
                                   feature_name_modifier = 'aisles',
                                   master_dataframe = df_master,
                                   users_dataframe = users)

result_1_departments = feature_generator(grouping_variable = 'department_id',
                                       feature_name_modifier = 'departments',
                                       master_dataframe = df_master,
                                       users_dataframe = users)
```

```python
def feature_generator_per_grouping_variable(g_id, grouping_variable,
                                            feature_name_modifier,
                                            master_dataframe = df_master):

    #Calculate total number of products ordered from an aisle
    num_products_from_grouping_variable = (master_dataframe[master_dataframe[grouping_variable] == g_id]
                                           .groupby('user_id')[grouping_variable]
                                           .count()
                                           .reset_index()
                                           .rename(columns = {grouping_variable :'num_products_from_'+feature_name_modifier+'_'+str(g_id)}))
    
    # Number of Orders that include atleast 1 product from the aisle
    number_of_orders_from_grouping_variable = (master_dataframe[master_dataframe[grouping_variable] == g_id]
                                               .groupby(['user_id'])['order_id']
                                               .count().reset_index()
                                               .rename(columns = {'order_id' :'number_of_orders_from_'+feature_name_modifier+'_'+str(g_id)}))

    # Number of Unique Products Ordered From Aisle
    num_unique_products_from_grouping_variable = (master_dataframe[master_dataframe[grouping_variable] == g_id]
                                                  .groupby('user_id')['product_id']
                                                  .nunique()
                                                  .reset_index()
                                                  .rename(columns = {'product_id' :'num_unique_products_ordered_from_'+feature_name_modifier+'_'+str(g_id)}))
    
    # Number of Unique Products Re-Oredered from Aisle
    num_unique_products_reordered_from_grouping_variable = (master_dataframe[(master_dataframe[grouping_variable] == g_id) & (master_dataframe['reordered'] == 1)]
                                                            .groupby('user_id')['product_id']
                                                            .nunique()
                                                            .reset_index()
                                                            .rename(columns = {'product_id' :'num_unique_products_reordered_from_'+feature_name_modifier+'_'+str(g_id)}))
    
    # Mean Number of Products Ordered From Aisle
    avg_num_of_products_ordered = (master_dataframe[master_dataframe[grouping_variable] == g_id]
                                  .groupby(['user_id','order_id'])[grouping_variable]
                                  .count()
                                  .reset_index()
                                  .groupby('user_id')
                                  .mean()
                                  .reset_index()
                                  .drop('order_id', axis = 1)
                                  .rename(columns = {grouping_variable:'avg_num_of_products_ordered_'+feature_name_modifier+'_'+str(g_id)}))
    
    # Mean Number of Products Re-Ordered From Aisle
    avg_num_of_products_reordered = (master_dataframe[(master_dataframe[grouping_variable] == g_id) & (master_dataframe['reordered'] == 1)]
                                    .groupby(['user_id','order_id'])[grouping_variable]
                                    .count()
                                    .reset_index()
                                    .groupby('user_id')
                                    .mean()
                                    .reset_index()
                                    .drop('order_id', axis = 1)
                                    .rename(columns = {grouping_variable:'avg_num_of_products_reordered_'+feature_name_modifier+'_'+str(g_id)}))
    
    # Std. Dev Number of Products Ordered From Aisle
    std_num_of_products_ordered = (master_dataframe[master_dataframe[grouping_variable] == g_id]
                                  .groupby(['user_id','order_id'])[grouping_variable]
                                  .count()
                                  .reset_index()
                                  .groupby('user_id')
                                  .std()
                                  .reset_index()
                                  .drop('order_id', axis = 1)
                                  .rename(columns = {grouping_variable:'std_num_of_products_ordered_'+feature_name_modifier+'_'+str(g_id)}))
    
    # Std. Dev Number of Products Re-Ordered From Aisle
    std_num_of_products_reordered = (master_dataframe[(master_dataframe[grouping_variable] == g_id) & (master_dataframe['reordered'] == 1)]
                                  .groupby(['user_id','order_id'])[grouping_variable]
                                  .count()
                                  .reset_index()
                                  .groupby('user_id')
                                  .std()
                                  .reset_index()
                                  .drop('order_id', axis = 1)
                                  .rename(columns = {grouping_variable:'std_num_of_products_reordered_'+feature_name_modifier+'_'+str(g_id)}))
    
    # Top 3 Product Ids and Counts ordered from each grouping variable (i.e aisle or department)
    frequency_of_products_ordered_from_grouping_variable = (master_dataframe[master_dataframe[grouping_variable] == g_id]
                                                            .groupby(['user_id', 'product_id'])['product_id']
                                                            .count()
                                                            .to_frame('count')
                                                            .reset_index()
                                                            .groupby(['user_id'])['product_id','count']
                                                            .apply(lambda grp: grp.nlargest(n = 3, columns = 'count'))
                                                            .reset_index())
    
    f = frequency_of_products_ordered_from_grouping_variable.groupby('user_id')['user_id'].count().to_frame('count')
    f.reset_index(inplace = True)
    # Users that have ordered atleast 3 items
    user_ids_3 = f[f['count'] == 3]['user_id']

    # Users that have ordered only 2 items
    user_ids_2 = f[f['count'] == 2]['user_id']

    # Users that have ordered only 1 item
    user_ids_1 = f[f['count'] == 1]['user_id']
    
    #Isolate the users by number of items
    frequency_of_products_ordered_from_grouping_variable_3items = frequency_of_products_ordered_from_grouping_variable[frequency_of_products_ordered_from_grouping_variable['user_id'].isin(user_ids_3)]
    frequency_of_products_ordered_from_grouping_variable_2items = frequency_of_products_ordered_from_grouping_variable[frequency_of_products_ordered_from_grouping_variable['user_id'].isin(user_ids_2)]
    frequency_of_products_ordered_from_grouping_variable_1items = frequency_of_products_ordered_from_grouping_variable[frequency_of_products_ordered_from_grouping_variable['user_id'].isin(user_ids_1)]

    # Add product_id and product_count column names
    frequency_of_products_ordered_from_grouping_variable_3items['product_id_column_names'] = [('num_1_most_frequent_product_id_from_'+feature_name_modifier+'_' + str(g_id)), ('num_2_most_frequent_product_id_from_'+feature_name_modifier+'_' + str(g_id)), ('num_3_most_frequent_product_id_from_'+feature_name_modifier+'_' + str(g_id))]*(len(user_ids_3))
    frequency_of_products_ordered_from_grouping_variable_3items['product_count_column_names'] = ['num_1_most_frequent_product_qty_from_'+feature_name_modifier+'_' + str(g_id), 'num_2_most_frequent_product_qty_from_'+feature_name_modifier+'_' + str(g_id), 'num_3_most_frequent_product_qty_from_'+feature_name_modifier+'_' + str(g_id)]*(len(user_ids_3))

    frequency_of_products_ordered_from_grouping_variable_2items['product_id_column_names'] = ['num_1_most_frequent_product_id_from_'+feature_name_modifier+'_' + str(g_id), 'num_2_most_frequent_product_id_from_'+feature_name_modifier+'_' + str(g_id)]*(len(user_ids_2))
    frequency_of_products_ordered_from_grouping_variable_2items['product_count_column_names'] = ['num_1_most_frequent_product_qty_from_'+feature_name_modifier+'_' + str(g_id), 'num_2_most_frequent_product_qty_from_'+feature_name_modifier+'_' + str(g_id)]*(len(user_ids_2))

    frequency_of_products_ordered_from_grouping_variable_1items['product_id_column_names'] = ['num_1_most_frequent_product_id_from_'+feature_name_modifier+'_' + str(g_id)]*(len(user_ids_1))
    frequency_of_products_ordered_from_grouping_variable_1items['product_count_column_names'] = ['num_1_most_frequent_product_qty_from_'+feature_name_modifier+'_' + str(g_id)]*(len(user_ids_1))

    # Drop unwanted columns
    frequency_of_products_ordered_from_grouping_variable_3items = frequency_of_products_ordered_from_grouping_variable_3items.drop([ 'product_id', 'count'], axis = 1)
    frequency_of_products_ordered_from_grouping_variable_2items = frequency_of_products_ordered_from_grouping_variable_2items.drop([ 'product_id', 'count'], axis = 1)
    frequency_of_products_ordered_from_grouping_variable_1items = frequency_of_products_ordered_from_grouping_variable_1items.drop([ 'product_id', 'count'], axis = 1)

    # Combine the three data frames
    frequency_of_products_ordered_from_grouping_variable_allitems = pd.concat([frequency_of_products_ordered_from_grouping_variable_3items, 
                                                                                frequency_of_products_ordered_from_grouping_variable_2items,
                                                                                frequency_of_products_ordered_from_grouping_variable_1items])
    
    # Left join with frequency_of_products_ordered dataframe to add column names to appropriate user ids
    frequency_of_products_ordered_from_grouping_variable = pd.merge(frequency_of_products_ordered_from_grouping_variable,
                                                                     frequency_of_products_ordered_from_grouping_variable_allitems,
                                                                     on = ['user_id', 'level_1'], how = 'left')
    
    top_3_product_ids_per_user_per_grouping_variable = frequency_of_products_ordered_from_grouping_variable.pivot(index = 'user_id',
                                                                 columns = 'product_id_column_names',
                                                                 values = 'product_id').reset_index()


    top_3_product_counts_per_user_per_grouping_variable = frequency_of_products_ordered_from_grouping_variable.pivot(index = 'user_id',
                                                                    columns = 'product_count_column_names',
                                                                    values = 'count').reset_index()

    list_of_feature_dfs_per_grouping_variable = [num_products_from_grouping_variable,
                                                 number_of_orders_from_grouping_variable,
                                                 num_unique_products_from_grouping_variable,
                                                 num_unique_products_reordered_from_grouping_variable,
                                                 avg_num_of_products_ordered,
                                                 avg_num_of_products_reordered,
                                                 std_num_of_products_ordered,
                                                 std_num_of_products_reordered,
                                                 top_3_product_ids_per_user_per_grouping_variable,
                                                 top_3_product_counts_per_user_per_grouping_variable]
    
    dfs_merged_per_grouping_variable = reduce(lambda  left,right: pd.merge(left,right,on='user_id',
                                            how='outer'), list_of_feature_dfs_per_grouping_variable)
    
    # Percentage of Unique Products Re-Ordered From Aisle
    percent_of_unique_products_reordered = pd.DataFrame()
    percent_of_unique_products_reordered['user_id'] = dfs_merged_per_grouping_variable.user_id
    percent_of_unique_products_reordered['percent_of_unique_products_reordered_from_'+feature_name_modifier+'_'+str(g_id)] = (dfs_merged_per_grouping_variable['num_unique_products_reordered_from_'+feature_name_modifier+'_'+str(g_id)]/
                                                                                                                              dfs_merged_per_grouping_variable['num_unique_products_ordered_from_'+feature_name_modifier+'_'+str(g_id)])
    dfs_merged_per_grouping_variable = pd.merge(dfs_merged_per_grouping_variable, percent_of_unique_products_reordered, on = 'user_id', how = 'left')
    
    return(dfs_merged_per_grouping_variable)
```

Since we have 134 aisles and 21 departments that we are generating features for, this process could be quite type consuming. To speed up the computation let's parallelize the process using python's multiprocessing library.

```python
# Generate Features for each aisle
# Step 1: Init multiprocessing.Pool()
pool = mp.Pool(mp.cpu_count())

# Step 2: `pool.apply` the `feature_generator_per_grouping_variable()`
list_results_aisles = pool.starmap(feature_generator_per_grouping_variable, [(g_id, 'aisle_id', 'aisles') for g_id in aisles.aisle_id])

# Step 3: Close the Pool
pool.close() 
```

```python
# Generate Features for each department
# Step 1: Init multiprocessing.Pool()
pool = mp.Pool(mp.cpu_count())

# Step 2: `pool.apply` the `feature_generator_per_grouping_variable()`
list_results_departments = pool.starmap(feature_generator_per_grouping_variable, [(g_id, 'department_id', 'departments') for g_id in departments.department_id])

# Step 3: Close the Pool
pool.close() 
```

Let's merge all the Aisle and Department specific features into the users dataframe

```python
results_aisles = reduce(lambda left,right: pd.merge(left,right,on='user_id',
                                                    how='outer'), list_results_aisles)
results_departments = reduce(lambda left,right: pd.merge(left,right,on='user_id',
                                                    how='outer'), list_results_departments)
```

```python
results_aisles = pd.merge(results_aisles, result_1_aisles, on = 'user_id', how = 'outer')
results_departments = pd.merge(results_departments, result_1_departments, on = 'user_id', how = 'outer')
```

```python
# Merge with users data
users = pd.merge(users, results_aisles, on = 'user_id', how = 'left')
users = pd.merge(users, results_departments, on = 'user_id', how = 'left')
```
