# Data Dictionary
`orders` (3.4m rows, 206k users):
* `order_id`: order identifier
* `user_id`: customer identifier
* `eval_set`: which evaluation set this order belongs in (see `SET` described below)
* `order_number`: the order sequence number for this user (1 = first, n = nth)
* `order_dow`: the day of the week the order was placed on
* `order_hour_of_day`: the hour of the day the order was placed on
* `days_since_prior`: days since the last order, capped at 30 (with NAs for `order_number` = 1)

`products` (50k rows):
* `product_id`: product identifier
* `product_name`: name of the product
* `aisle_id`: foreign key
* `department_id`: foreign key

`aisles` (134 rows):
* `aisle_id`: aisle identifier
* `aisle`: the name of the aisle

`deptartments` (21 rows):
* `department_id`: department identifier
* `department`: the name of the department

`order_products__SET` (30m+ rows):
* `order_id`: foreign key
* `product_id`: foreign key
* `add_to_cart_order`: order in which each product was added to cart
* `reordered`: 1 if this product has been ordered by this user in the past, 0 otherwise

where `SET` is one of the four following evaluation sets (`eval_set` in `orders`):
* `"prior"`: orders prior to that users most recent order (~3.2m orders)
* `"train"`: training data supplied to participants (~131k orders)
* `"test"`: test data reserved for machine learning competitions (~75k orders)

# Primary Objective

To predict how likely a customer(identified by user_id) is to re-order each of the top X products(identified by product_id) from the ____ department(identified by department_id)

### Most common words in product names

```python
# Convert the names to lower case, split the names by whitespace,
# stack the resulting series and then count the frequency
df_product_name_word_counts = products.product_name.str.lower().str.split(expand = True).stack().value_counts()

# convert to a dataframe and rename the column
df_product_name_word_counts = df_product_name_word_counts.to_frame('counts')

# adjust the index
df_product_name_word_counts.reset_index(inplace = True)

# Rename the column  "index" to "word"
df_product_name_word_counts = df_product_name_word_counts.rename(columns = {'index' : 'word'})

# Drop stop words like (with, and, the, etc) using nltk's stopword list
df_product_name_word_counts = df_product_name_word_counts[~df_product_name_word_counts['word'].isin(stopwords.words('english'))]

# Drop any special characters or punctuation
df_product_name_word_counts = df_product_name_word_counts[~df_product_name_word_counts['word'].isin(set(punctuation))]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,6))
sns.barplot(ax = ax, data =df_product_name_word_counts[0:50],  x = "word", y = "counts", alpha = 0.5, ci = None)
plt.xticks(rotation=45)
plt.title('50 Most Frequency Words in Product Names')
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.show()
```

* The number of products in an aisle ranges from 12 to 1258 with the average number of product in an aisle being 370

* Organic seems to be the most common word in product names and leads me to believe that insta cart sells a lot of organic products. Chocolate, Cheese, Chicken and Original are also among the most common words in product names  
* The instacart data base has users that have ordered as few as three times from the platform, and as much as a 100 times. However the distribution of the orders is heavily skewed to the right, with most people having ordered up to 40 times or less. While we are unaware of the time duration of this data, it would be interesting to explore why customers order more or less often.  
* The orders are evenly distributed over the week, except a slight increase over the weekends(Saturday and Sunday), which would be expected.  
* Furthermore, a bulk of there orders come between 8am and 10pm and follow the same pattern for every day of the week  
* The data is censored at 30+ and there are a large number of users that are reported as ordering at the 30 day mark. However, since censoring at 30 implies that users that ordered at the 30 day mark or even after are treated the same way and recorded in the data as 30. This creates an artificial spike in the data that does not add any interpretive value.  
* The distribution of the number of days before people re-order is skewed to the right with a majority or the re-orders coming between 0-15 days from their previous order. This would make sense if the quantities of the things people order would usually be consumed within 15 days of them buying it.  
* Furthermore, there are also spikes in the data around the 7, 14, 21 and 28 day mark that diverge from the trend of the data. This could potentially represent peoples habits of shopping for an item on a weekly, bi-weekly, tri-weekly or monthly basis.   
* Looking at the orders, it appears that Organic products are heavily featured amongst the re-ordered items, with Bag of Organic Bananas, Organic  Strawberries and Organic Baby Spinach being among the  top 5 most re-ordered products.  
* The aisles for Fresh Fruits and Fresh Vegetables account for the most number of re-orders  
* Amongst the departments, Produce is the one that is most frequently re-ordered from, with dairy and eggs being  second.  

# Data Engineering
It seems that missing from our data is a table for users. Considering that our primary research question is trying to predict the behavior of a user (i.e. whether or not they will re_order certain items), I think it is imperative to create a table for users. We will use this same table to populate features that we engineer for each user. 

```python
users = pd.DataFrame(orders_non_test['user_id'].unique(), columns=['user_id'])
users.head()
```

Furthermore, since the data is distributed amongst many tables, I foresee many instances of inner and left joins which are computationally expensive task, and hence I'm gonna create a master table with all the relevant data combined into a single dataframe. The downside to this being that this dataframe will have redundant information with repeating values. This is why data is usually stored in multiple tables with and entity relationship between the tables.

```python
a1 = pd.merge(orders_prior, order_products_prior, on = 'order_id', how = 'left')
a2 = pd.merge(a1, products, on = 'product_id', how = 'left')
a3 = pd.merge(a2, aisles, on = 'aisle_id', how = 'left')
df_master = pd.merge(a3, departments, on = 'department_id', how = 'left')
df_master.head()
```

Let's start adding some features to the users table

## Feature Engineering

### Count of Number Orders per User
The total number of orders (prior) that each user has.

```python
df_order_count_per_user = df_master.groupby(['user_id'])['order_id'].nunique().to_frame('counts')
df_order_count_per_user.reset_index(inplace = True)
df_order_count_per_user.head()
```

```python
df_order_count_per_user = orders_prior.groupby('user_id')['user_id'].count().to_frame('number_of_orders')
df_order_count_per_user.reset_index(inplace = True)
users = pd.merge(users, df_order_count_per_user, on = 'user_id', how = 'inner')
users.head()
```

### Number of Products Ordered by each User (prior) 
The total number of products ordered by each user prior to their most recent order.This is feature is not showing you the total number of unique products ordered. 

```python
products_per_order = order_products_prior.groupby('order_id')['order_id'].count().to_frame('number_of_products_ordered')
products_per_order.reset_index(inplace = True)
products_per_order.head()
```

```python
products_per_order = pd.merge(products_per_order, orders_prior, on = 'order_id', how = 'left')
products_per_user = products_per_order.groupby('user_id')['number_of_products_ordered'].sum().to_frame('total_products_ordered_prior')
products_per_user.reset_index(inplace = True)
users = pd.merge(users, products_per_user, on = 'user_id', how = 'left')
users.head()
```

### Number of Unique Products ordered (prior)
The number unique products ordered by the user over his/her entire shopping history with instacart

```python
order_products_prior.head()
```

```python
unique_products_per_user = df_master.groupby('user_id')['product_id'].nunique().to_frame('number_of_unique_products_ordered')
unique_products_per_user.reset_index(inplace = True)
users = pd.merge(users, unique_products_per_user, on = 'user_id', how = 'left')
users.head()
```

### Number of Products Re-Ordered
This feature represents the total number of products ordered that were re-orders by the user after the first order.

```python
products_reordered_per_user = df_master.groupby('user_id')['reordered'].sum().to_frame('total_products_reordered')
products_reordered_per_user.reset_index(inplace = True)
users = pd.merge(users, products_reordered_per_user, on = 'user_id', how = 'left')
users.head()
```

### Number of Unique Products Re-Ordered at least once
This feature represents the total number of unique products that were re-ordered by the user after the first order.

```python
unique_products_reordered_per_user = df_master[df_master['reordered'] == 1].groupby('user_id')['product_id'].nunique().to_frame('number_of_unique_products_reordered')
unique_products_reordered_per_user.reset_index(inplace = True)
users = pd.merge(users, unique_products_reordered_per_user, on = 'user_id', how = 'left')
users.head()
```

### Mean Number of Products per Order
This feature represents the average number of products in each order

```python
users['avg_products_per_order'] = users['total_products_ordered_prior'] / users['number_of_orders']
users.head()
```

### Mode Number of Products per Order
What is the mode (or most frequent) number of products the user orders per order

```python
mode_products_per_order = products_per_order.groupby('user_id')['number_of_products_ordered'].agg(pd.Series.mode).to_frame('mode_products_ordered')
mode_products_per_order.reset_index(inplace = True)
mode_products_per_order['mode_products_ordered'] = mode_products_per_order['mode_products_ordered'].map(lambda x: x.mean())
users = pd.merge(users, mode_products_per_order, on = 'user_id', how = 'left')
users.head()
```

### Standard Deviation of Number of Products per Order
This feature will represetnt the statndard deviation of the number of products ordered a user in a single order

```python
std_dev_products_per_order = products_per_order.groupby('user_id')['number_of_products_ordered'].std().to_frame('std_dev_products_ordered')
std_dev_products_per_order.reset_index(inplace = True)
users = pd.merge(users, std_dev_products_per_order, on = 'user_id', how = 'left')
users.head()
```

### Percentage of products re-ordered
The percentage of the unique products - ordered by the user - that were re-ordered

```python
users['percent_unique_products_reordered'] = users['number_of_unique_products_reordered'] / users['number_of_unique_products_ordered']
users.head()
```

### Top 3 Most Frequently bought Product ids and the counts
Engineering 6 features in total. The top 3 Most frequently bought product ids and the respective counts representing how many times each product was bough. A value of NaN in any of the features suggests that the user had bought less than 3 items. So for instance if the the 3rd most frequently bought product id and count is NaN, it implies that the user only ever purchased 2 items from insta-cart

```python
frequency_of_products_ordered = df_master.groupby(['user_id', 'product_id'])['product_id'].count().to_frame('count')
frequency_of_products_ordered.reset_index(inplace = True)
frequency_of_products_ordered = frequency_of_products_ordered.groupby(['user_id'])['product_id','count'].apply(lambda grp: grp.nlargest(n = 3, columns = 'count'))
frequency_of_products_ordered.reset_index(inplace = True)
frequency_of_products_ordered.head()
```

```python
f1 = frequency_of_products_ordered.groupby('user_id')['user_id'].count().to_frame('count')
f1.reset_index(inplace = True)
# Users that have ordered atleast 3 items
user_ids_3 = f1[f1['count'] == 3]['user_id']

# Users that have ordered only 2 items
user_ids_2 = f1[f1['count'] == 2]['user_id']

# Users that have ordered only 1 item
user_ids_1 = f1[f1['count'] == 1]['user_id']
```

```python
frequency_of_products_ordered.head()
```

```python
#Isolate the users by number of items
frequency_of_products_ordered_3items = frequency_of_products_ordered[frequency_of_products_ordered['user_id'].isin(user_ids_3)]
frequency_of_products_ordered_2items = frequency_of_products_ordered[frequency_of_products_ordered['user_id'].isin(user_ids_2)]
frequency_of_products_ordered_1items = frequency_of_products_ordered[frequency_of_products_ordered['user_id'].isin(user_ids_1)]

# Add product_id and product_count column names
frequency_of_products_ordered_3items['product_id_column_names'] = ['num_1_most_frequent_product_id', 'num_2_most_frequent_product_id', 'num_3_most_frequent_product_id']*(len(user_ids_3))
frequency_of_products_ordered_3items['product_count_column_names'] = ['num_1_most_frequent_product_qty', 'num_2_most_frequent_product_qty', 'num_3_most_frequent_product_qty']*(len(user_ids_3))

frequency_of_products_ordered_2items['product_id_column_names'] = ['num_1_most_frequent_product_id', 'num_2_most_frequent_product_id']*(len(user_ids_2))
frequency_of_products_ordered_2items['product_count_column_names'] = ['num_1_most_frequent_product_qty', 'num_2_most_frequent_product_qty']*(len(user_ids_2))

frequency_of_products_ordered_1items['product_id_column_names'] = ['num_1_most_frequent_product_id']*(len(user_ids_1))
frequency_of_products_ordered_1items['product_count_column_names'] = ['num_1_most_frequent_product_qty']*(len(user_ids_1))


# Drop unwanted columns
frequency_of_products_ordered_3items = frequency_of_products_ordered_3items.drop([ 'product_id', 'count'], axis = 1)
frequency_of_products_ordered_2items = frequency_of_products_ordered_2items.drop([ 'product_id', 'count'], axis = 1)
frequency_of_products_ordered_1items = frequency_of_products_ordered_1items.drop([ 'product_id', 'count'], axis = 1)

# Combine the three data frames
frequency_of_products_ordered_allitems = pd.concat([frequency_of_products_ordered_3items, 
                                                    frequency_of_products_ordered_2items,
                                                    frequency_of_products_ordered_1items])


# Left join with frequency_of_products_ordered dataframe to add column names to appropriate user ids
frequency_of_products_ordered = pd.merge(frequency_of_products_ordered,
                                         frequency_of_products_ordered_allitems,
                                         on = ['user_id', 'level_1'], how = 'left')

frequency_of_products_ordered.head()
```

```python
top_3_product_ids_per_user = frequency_of_products_ordered.pivot(index = 'user_id',
                                                                 columns = 'product_id_column_names',
                                                                 values = 'product_id')
top_3_product_ids_per_user.reset_index(inplace = True)
users = pd.merge(users, top_3_product_ids_per_user, on = 'user_id', how = 'left')
```

```python
top_3_product_counts_per_user = frequency_of_products_ordered.pivot(index = 'user_id',
                                                                    columns = 'product_count_column_names',
                                                                    values = 'count')
top_3_product_counts_per_user.reset_index(inplace = True)
users = pd.merge(users, top_3_product_counts_per_user, on = 'user_id', how = 'left')
```
